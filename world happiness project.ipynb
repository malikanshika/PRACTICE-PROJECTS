```python
X_train,X_test,Y_train,Y_test=train_test_split(feature_vars,target_var,train_size=.75)
lm=linear_model.LinearRegression()
lm.fit(X_train,Y_train)
```

```python
r2_score(Y_test,y_pred)
```
